```python
# import libraries
import pandas as pd
```

```python
# import data
responses = pd.DataFrame(pd.read_csv("data/Random Sample of Data files/responses.csv"))
```

```python
# SEPARATING DATASETS 

# filter out code questions
code_responses = responses[responses["item_type"].str.contains("code")]

# # separate the responses into their respective textbooks and latest versions??
# code_responses_ABCD = code_responses[code_responses["book"].str.contains("ABCD")]
# code_responses_ABC = code_responses[~code_responses["book"].str.contains("ABCD")]

# code_responses_ABC_College = code_responses_ABC[code_responses_ABC["book"].str.contains("College")]
# code_responses_ABC_HighSchool = code_responses_ABC[code_responses["book"].str.contains("High School")]

# filter out learnosity questions
learnosity_responses = responses[responses["item_type"].str.contains("learnosity")]

# # separate the responses into their respective textbooks and latest versions??
# learnosity_responses_ABCD = learnosity_responses[learnosity_responses["book"].str.contains("ABCD")]
# learnosity_responses_ABC = learnosity_responses[~learnosity_responses["book"].str.contains("ABCD")]

# learnosity_responses_ABC_College = learnosity_responses_ABC[learnosity_responses_ABC["book"].str.contains("College")]
# learnosity_responses_ABC_HighSchool = learnosity_responses_ABC[learnosity_responses["book"].str.contains("High School")]
```

```python


# Filter out attempts where the student got the question wrong
correct_attempts = responses[responses['points_earned'] == responses['points_possible']]

# Find the largest and smallest attempt numbers where the student got the question right
min_max_attempts = correct_attempts.groupby(['student_id', 'item_id'])['attempt'].agg(['min', 'max']).reset_index()

# Calculate the number of review attempts
min_max_attempts['review_attempts'] = min_max_attempts['max'] - min_max_attempts['min']

# Pivot the data to have columns for each question with the number of review attempts
pivoted_data = min_max_attempts.pivot(index='student_id', columns='item_id', values='review_attempts').reset_index()

pivoted_data["total_review_attempts"] = pivoted_data.iloc[:, 1:].sum(axis=1)

# Display the transformed dataset
display(pivoted_data)
```

```python
code_review_attempts_per_chapter = {}

# Loop through unique chapters and create pivoted_data for each chapter
for chapter in code_responses['chapter'].unique():
    # Filter responses for the current chapter
    chapter_responses = code_responses[code_responses["chapter"] == chapter]
    
    # Filter out attempts where the student got the question wrong
    correct_attempts = chapter_responses[chapter_responses['points_earned'] == chapter_responses['points_possible']]
    
    # Find the largest and smallest attempt numbers where the student got the question right
    min_max_attempts = correct_attempts.groupby(['student_id', 'item_id'])['attempt'].agg(['min', 'max']).reset_index()
    
    # Calculate the number of review attempts
    min_max_attempts['review_attempts'] = min_max_attempts['max'] - min_max_attempts['min']
    
    # Pivot the data to have columns for each question with the number of review attempts
    pivoted_data = min_max_attempts.pivot(index='student_id', columns='item_id', values='review_attempts').reset_index()
    
    # Add a new column "total_review_attempts"
    pivoted_data["total_review_attempts"] = pivoted_data.iloc[:, 1:].sum(axis=1)
    
    # Store the result in the dictionary
    review_attempts_per_chapter[chapter] = pivoted_data

# # Display the pivoted_data for each chapter
# for chapter, df in review_attempts_per_chapter.items():
#     print(f"Chapter: {chapter}")
#     display(df)
```
