```python
# import libraries
import pandas as pd
```

```python
# import data
responses = pd.DataFrame(pd.read_csv("data/Random Sample of Data files/responses.csv"))
```

```python
# SEPARATING DATASETS 

# filter out code questions
code_responses = responses[responses["item_type"].str.contains("code")]

# separate the responses into their respective textbooks and latest versions??
code_responses_ABCD = code_responses[code_responses["book"].str.contains("ABCD")]
code_responses_ABC = code_responses[~code_responses["book"].str.contains("ABCD")]

code_responses_ABC_College = code_responses_ABC[code_responses_ABC["book"].str.contains("College")]
code_responses_ABC_HighSchool = code_responses_ABC[code_responses["book"].str.contains("High School")]

# filter out learnosity questions
learnosity_responses = responses[responses["item_type"].str.contains("learnosity")]

# separate the responses into their respective textbooks and latest versions??
learnosity_responses_ABCD = learnosity_responses[learnosity_responses["book"].str.contains("ABCD")]
learnosity_responses_ABC = learnosity_responses[~learnosity_responses["book"].str.contains("ABCD")]

learnosity_responses_ABC_College = learnosity_responses_ABC[learnosity_responses_ABC["book"].str.contains("College")]
learnosity_responses_ABC_HighSchool = learnosity_responses_ABC[learnosity_responses["book"].str.contains("High School")]
```

```python
# ALL STUDENTS AVERAGE ATTEMPTS FOR CODE PROBLEMS
# All students (regardless of different book)
correct_responses = code_responses[code_responses["points_possible"] == code_responses["points_earned"]]

# Find the attempt with the least number of attempts where the student got the question right
code_min_attempts = correct_responses.groupby(['student_id', 'item_id'])['attempt'].min().reset_index()

# pivot the data
code_min_attempts = code_min_attempts.pivot(index='student_id', columns='item_id', values='attempt').reset_index()


code_min_attempts = code_min_attempts.fillna(0)

code_min_attempts['average_attempts'] = code_min_attempts.iloc[:, 1:].mean(axis=1)

display(code_min_attempts)
print(len(code_min_attempts))
```

```python
# CODE ATTEMPTS FOR EVERY STUDENT PER CHAPTER

code_min_attempts_per_chapter = {}

code_correct_responses = code_responses[code_responses["points_possible"] == code_responses["points_earned"]]

for chapter in code_responses['chapter'].unique():
    # Filter responses for the current chapter
    chapter_responses = code_correct_responses[code_correct_responses["chapter"] == chapter]
    
    # Find the attempt with the least number of attempts where the student got the question right
    chapter_code_min_attempts = chapter_responses.groupby(['student_id', 'item_id'])['attempt'].min().reset_index()
    
    # Pivot the data
    chapter_code_min_attempts = chapter_code_min_attempts.pivot(index='student_id', columns='item_id', values='attempt').reset_index()
    
    # Fill NaN values with 0
    chapter_code_min_attempts = chapter_code_min_attempts.fillna(0)
    
    # Calculate the average attempts
    chapter_code_min_attempts['average_attempts'] = chapter_code_min_attempts.iloc[:, 1:].mean(axis=1)
    
    # Store the result in the dictionary
    code_min_attempts_per_chapter[chapter] = chapter_code_min_attempts

# # Display the code_min_attempts for each chapter
# for chapter, df in code_min_attempts_per_chapter.items():
#     print(f"Chapter: {chapter}")
#     display(df)
```
