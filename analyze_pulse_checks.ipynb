```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('random subset of data files/checkpoints.csv')
```

```python
cost = df[df['construct'] == "Cost"]
cost_nna = cost.dropna(subset=['response'])
#cost_nna.head()
```

```python
expect = df[df['construct'] == "Expectancy"]
expect_nna = expect.dropna(subset=['response'])
#expect_nna.head()
```

```python
intrins = df[df['construct'] == "Intrinsic Value"]
intrins_nna = intrins.dropna(subset=['response'])
#intrins_nna.head()
```

```python
util = df[df['construct'] == "Utility Value"]
util_nna = util.dropna(subset=['response'])
#util_nna.head()
```

```python
# Now, we'll create the plot using seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=cost_nna, x='response', y='EOC', marker='o')

plt.title('EOC Scores by Cost Pulse Response for All Students')
plt.xlabel('Cost Response Score')
plt.ylabel('EOC Score')
plt.legend(title='Student ID', bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
# Now, we'll create the plot using seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=expect_nna, x='response', y='EOC', marker='o')

plt.title('EOC Scores by Expectancy Pulse Response for All Students')
plt.xlabel('Expectancy Response Score')
plt.ylabel('EOC Score')
plt.legend(title='Student ID', bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
# Now, we'll create the plot using seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=intrins_nna, x='response', y='EOC', marker='o')

plt.title('EOC Scores by Intrinsic Value Pulse Response for All Students')
plt.xlabel('Intrinsic Value Response Score')
plt.ylabel('EOC Score')
plt.legend(title='Student ID', bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
# Now, we'll create the plot using seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=util_nna, x='response', y='EOC', marker='o')

plt.title('EOC Scores by Utility Value Pulse Response for All Students')
plt.xlabel('Utility Value Response Score')
plt.ylabel('EOC Score')
plt.legend(title='Student ID', bbox_to_anchor=(1.05, 1), loc='upper left')
```
